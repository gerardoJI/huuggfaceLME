```python
import PyPDF2

def extraer_texto_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        texto = ""
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            texto += page.extract_text()
        return texto

texto = extraer_texto_pdf("pdfs\Jiménez Gerardo DRM-AE3.pdf")
print(texto)
```

```python
import subprocess

# Ruta completa al ejecutable de Ollama
OLLAMA_PATH = r"C:\Users\Gerardo\AppData\Local\Programs\Ollama\ollama.exe"

def query_ollama_run(model, prompt):
    try:
        # Iniciar el proceso interactivo con `ollama run`
        process = subprocess.Popen(
            [OLLAMA_PATH, "run", model],
            stdin=subprocess.PIPE,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
        # Enviar el mensaje al modelo
        stdout, stderr = process.communicate(input=prompt)
        return stdout.strip() if stdout else stderr.strip()
    except Exception as e:
        return f"Error al ejecutar Ollama: {e}"

prompt_resumen = f"Traduce a inglés y resume el siguiente texto de manera concisa y clara, sin numerar: {texto}"

# Prueba con el modelo Phi
respuesta = query_ollama_run("phi", prompt_resumen)
print("Respuesta de Phi:", respuesta)
```
